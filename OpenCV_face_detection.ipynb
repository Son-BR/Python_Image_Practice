# OpenCV 얼굴 탐지

```python
# packages
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
```

```python
# 이미지 파일 불러와서 얼굴 탐지 후 원본과 얼굴탐지 이미지 출력 함수

def face_detect(path):
    
    
    # 이미지 파일 읽기
    im = cv2.imread(path) 
    
    # 원본 이미지 파일 컬러 변경
    fix_img2 = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) # 컬러 채널 변경
    
    # 이미지 원본 보여줌
    plt.figure(figsize=(15,10))
    plt.subplot(121)
    plt.imshow(fix_img2)
    plt.axis("off")
    # plt.show()
    
    # detect faces (얼굴 검출)
    faces, confidences = cv.detect_face(im)
    
    # 얼굴 검출 후 컬러변경(컬러 회색일때 얼굴 검출이 더 잘됨)
    fix_img = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) # 컬러 채널 변경
    
    # loop through detected faces and add bounding box (검출된 얼굴을 박스로 감싸기)
    for face in faces:
        (startX,startY) = face[0],face[1]
        (endX,endY) = face[2],face[3]
        # draw rectangle over face
        cv2.rectangle(fix_img, (startX,startY), (endX,endY), (0,255,0), 2)
    
    # 얼굴검출 후 박스로 감싼 이미지 출력
    plt.subplot(122)  
    plt.imshow(fix_img)
    plt.axis("off")
    plt.show()

    # 이미지 저장
    # cv2.imwrite('result.jpg', im) # 이미지 쓰기
    
```

```python
# 폴더 내에 있는 이미지 전부 적용

import os
```

```python
image_list=os.listdir('./Data/Image')

for i in image_list:
    if i[-3:] == 'jpg' or i[-3:] == 'png':
        face_detect(f'./Data/Image/{i}')
```

```python
# 이미지 보여주기
image_path = './Data/Image/02.jpg'
im = cv2.imread(image_path) # 이미지 읽기

fix_img = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

# 새창에서 띄우기
# cv2.imshow('fix_img', im)
# cv2.waitKey(0)

plt.imshow(fix_img)
plt.axis('off')
plt.show()
```

```python
# detect faces (얼굴 검출)
faces, confidences = cv.detect_face(im)
```

```python
# loop through detected faces and add bounding box (검출된 얼굴을 박스로 감싸기)
for face in faces: # faces : 얼굴이 여러개 일때 
    (startX,startY) = face[0],face[1]
    (endX,endY) = face[2],face[3]
    # draw rectangle over face
    cv2.rectangle(im, (startX,startY), (endX,endY), (0,255,0), 2)
```
