```python
import torch

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Images
imgs = ['https://ultralytics.com/images/zidane.jpg']  # batch of images

# Inference
results = model(imgs)

# Results
results.print()
results.save()  # or .show()

results.xyxy[0]  # img1 predictions (tensor)
results.pandas().xyxy[0]  # img1 predictions (pandas)

results.show()
```

```python
# packages
import torch
import cvlib as cv
import cv2
import os
```

```python
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5n', pretrained=True)

webcam=cv2.VideoCapture(0)

if not webcam.isOpened():
    print("Could not open webcam")
    exit()
    
# loop through frames
while webcam.isOpened():
 
    # read frame from webcam 
    status, frame = webcam.read()
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
 
    if not status:
        print("Could not read frame")
        exit()
   
    results = model(frame)

    # display output
    # results.show()
 
    # press "Q" to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# release resources
webcam.release()
cv2.destroyAllWindows()

print('Finish')


# im=cv2.imread(file_path)  # 이미지 읽기
# im=cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
import torch
import cv2

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5n', pretrained=True)

webcam=cv2.VideoCapture("./Data/Video/humans_pixabay.mp4")

if not webcam.isOpened():
    print("Could not open webcam")
    exit()
    
# loop through frames
while webcam.isOpened():
 
    # read frame from webcam 
    status, frame = webcam.read()
    frame=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
 
    if not status:
        print("Could not read frame")
        exit()
    
    results = model(frame)
    results.save()
    
    img = cv2.imread('./runs/detect/exp/image0.jpg') # cv2.IMREAD_REDUCED_COLOR_2
    cv2.imshow('img', img)

    if os.path.isfile('./runs/detect/exp/image0.jpg'):
        os.remove('./runs/detect/exp/image0.jpg')
        os.rmdir('./runs/detect/exp/')
    
 
    # press "Q" to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# release resources
webcam.release()
cv2.destroyAllWindows()

print('Finish')


# im=cv2.imread(file_path)  # 이미지 읽기
# im=cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```
